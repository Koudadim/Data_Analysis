### Classification des étudiants

```python
# ajouter les trois colonnes fixe_score, croissance_score et mindset
mindset_data['fixe_score'] = [''] * 128
mindset_data['croissance_score'] = [''] * 128
mindset_data['mindset'] = [''] * 128

mindset_data.head()
```

### La fonction qui classifie les étudiants

```python
ourReference = {
    "Q1" : "fixe",
    "Q2" : "croissance",
    "Q3" : "fixe",
    "Q4" : "fixe",
    "Q5" : "fixe",
    "Q6" : "croissance",
    "Q7" : "fixe",
    "Q8" : "croissance",
    "Q9" : "fixe",
    "Q10" : "croissance",    
}
def classer_mentalite(index):

    print('classify student ',index, '...')

    esprit_fixe_score = 0
    esprit_croissance_score = 0

    for i in range(1,11):
        column = 'Q'+str(i)
        response = mindset_data[column][index]
        ref = ourReference[column]
        if response == 1 :
            if ref == "croissance" :
                esprit_croissance_score += 1
            elif ref == "fixe" :
                esprit_fixe_score += 1
        elif response == 0 :
            if ref == "croissance" :
                esprit_fixe_score += 1
            elif ref == "fixe" :
                esprit_croissance_score += 1

    mindset_data["croissance_score"][index] = esprit_croissance_score
    mindset_data["fixe_score"][index] = esprit_fixe_score

    if esprit_fixe_score > esprit_croissance_score:
        mindset_data["mindset"][index] = "fixe"
    elif esprit_croissance_score > esprit_fixe_score:
        mindset_data["mindset"][index] = "croissance"
    else:
        mindset_data["mindset"][index] = "Indéterminé"

    print('done')
for id in range(0,128):
    classer_mentalite(id)
    print("---------------------")
```

### Appliquer l'AFD

```python
from sklearn import datasets
import pandas as pd
import numpy as np
np.set_printoptions(precision=4)
from matplotlib import pyplot as plt
import seaborn as sns
sns.set()
from sklearn.preprocessing import LabelEncoder
```

### Définir les variables

```python
X = mindset_data.iloc[:, :10]
y = mindset_data.iloc[:, -1]
```

```python
X.head(2)
```

```python
y.head(2)
```

```python
# Ajouter la colonne class à X et les mettre dans un dataframe
df = X.join(pd.Series(y, name='mindset'))
df.head()
```

```python
# calculer la matrice des centres de gravité des groupes
target_names = list(df['mindset'].unique())

class_feature_means = pd.DataFrame(columns=target_names)
for c, rows in df.groupby('mindset'):
    class_feature_means[c] = rows.mean()
class_feature_means
```

```python
# Matrice de variance intra_classes (la matrice W)

within_class_scatter_matrix = np.zeros((10,10))
for c, rows in df.groupby('mindset'):
    rows = rows.drop(['mindset'], axis=1)
    
    s = np.zeros((10,10))
    for index, row in rows.iterrows():
        x, mc = row.values.reshape(10,1), class_feature_means[c].values.reshape(10,1)
        
        s += (x - mc).dot((x - mc).T)
    
    within_class_scatter_matrix += s
```

```python
# Matrice de variance inter_classes (la matrice B)
feature_means = df.mean()
between_class_scatter_matrix = np.zeros((10,10))
for c in class_feature_means:    
    n = len(df.loc[df['mindset'] == c].index)
    
    mc, m = class_feature_means[c].values.reshape(10,1), feature_means.values.reshape(10,1)
    
    between_class_scatter_matrix += n * (mc - m).dot((mc - m).T)
```

```python
# calcul des valuers et vecteurs propres 
eigen_values, eigen_vectors = np.linalg.eig(np.linalg.inv(within_class_scatter_matrix).dot(between_class_scatter_matrix))
```

```python
# Tris des valeurs propres
# la liste pair contient les pairs (valeur propre et vecteur propre coresspondant)
pairs = [(np.abs(eigen_values[i]), eigen_vectors[:,i]) for i in range(len(eigen_values))]
# trier le liste
pairs = sorted(pairs, key=lambda x: x[0], reverse=True)
# afficher les valeurs propres
for pair in pairs:
    print(pair[0])
```

```python
# Ratio des valeurs propres
eigen_value_sums = sum(eigen_values)
print('Explained Variance')
for i, pair in enumerate(pairs):
    print('Eigenvector {}: {}'.format(i, (pair[0]/eigen_value_sums).real))
```

```python
# Vecteurs propres des deux premières dimensions
w_matrix = np.hstack((pairs[0][1].reshape(10,1), pairs[1][1].reshape(10,1))).real
```

```python
# Nouvelles coordonnées (X . w)
X_lda = np.array(X.dot(w_matrix))
```

```python
# Codage des labels des classes
le = LabelEncoder()
# ici en transforme les valeurs de la colonne class en nombre entiers
y = le.fit_transform(df['mindset'])
```

```python
# la visualisation du nuage de points en utilisant les valeurs transformées de la colonne de mindset.
plt.xlabel('LD1')
plt.ylabel('LD2')
plt.scatter(
    X_lda[:,0],
    X_lda[:,1],
    c=y,
    cmap='rainbow',
    alpha=0.7,
    edgecolors='b'
)
```
