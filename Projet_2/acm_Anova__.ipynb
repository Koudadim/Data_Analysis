## Application d'ANOVA à un facteur pour voir l'impact de la durée du trajet sur la présence de l'étudiant au cours.

```python
data_anova = data[["Q3","Q7","Q8"]]
data_anova
```

**Encodage de la variable Q8**

```python
from sklearn.preprocessing import LabelEncoder

# Création d'une instance de LabelEncoder
label_encoder = LabelEncoder()

# Encodage de la variable Q8
data_anova['Q8_encoded'] = label_encoder.fit_transform(data_anova['Q8'])
```

```python
data_anova
```

**Vérification de dépendance entre les variables Q7 et Q8.
Q7 = 'Quelle est la durée de trajet de votre domicile à la faculté en minutes ?'
Q8 = 'Vous assistez aux cours…'**

```python
data_anova['Q7'].unique()
```

```python
from scipy.stats import chi2_contingency

# Création d'une table de contingence entre les colonnes Q7 et Q8
contingency_table = pd.crosstab(data_anova['Q7'], data_anova['Q8_encoded'])

# Application du test du Chi-square
chi2, p_value, _, _ = chi2_contingency(contingency_table)

# Affichage des résultats
print("Test du Chi-square")
print("Chi2 :", chi2)
print("p-value :", p_value)
```

**Interprétation :**
La statistique Chi2 mesure la déviation entre les observations observées et les valeurs attendues sous l'hypothèse nulle d'indépendance entre les variables "Q7" et "Q8". Une valeur plus élevée de la statistique Chi2 indique une plus grande divergence par rapport à l'indépendance.

La valeur p est une mesure de la significativité statistique. Elle indique la probabilité d'observer une statistique Chi2 aussi extrême que celle calculée, sous l'hypothèse nulle d'indépendance réelle entre les variables. Une valeur p faible (inférieure à un seuil prédéfini, généralement 0,05) suggère un rejet de l'hypothèse nulle d'indépendance et indique une association significative entre les variables.

Dans notre cas, avec une valeur p de 0.006, qui est inférieure à 0.05, on peut conclure qu'il existe une association significative entre les variables "Q7" et "Q8". Cela signifie que les catégories des variables "Q7" et "Q8" ne sont pas indépendantes et qu'il y a une relation statistiquement significative entre elles.

**Vérification de dépendance entre les variables Q2 et Q8. Q2 = 'Votre formation ?' Q8 = 'Vous assistez aux cours…'**

```python
from scipy.stats import chi2_contingency

# Création d'une table de contingence entre les colonnes Q7 et Q8
contingency_table = pd.crosstab(data_anova['Q3'], data_anova['Q8_encoded'])

# Application du test du Chi-square
chi2, p_value, _, _ = chi2_contingency(contingency_table)

# Affichage des résultats
print("Test du Chi-square")
print("Chi2 :", chi2)
print("p-value :", p_value)
```

Ici, on a une valeur p de 0.0003, qui est inférieure à 0.05, on peut conclure qu'il existe une association significative entre les variables "Q3" et "Q8". Cela signifie que les catégories des variables "Q3" et "Q8" ne sont pas indépendantes et qu'il y a une relation statistiquement significative entre elles.

A l’aide de la fonction anova_lm, on lance une ANOVA à un facteur en considérant la variable Q8 et une variable explicative Q7 et Q3 à la fois

```python
from statsmodels.formula.api import ols
import statsmodels.api as sm


model = ols('Q8_encoded ~ C(Q7)', data=data_anova).fit()
anova_table = sm.stats.anova_lm(model, typ=1)
anova_table
```

**Interprétation :** Ici, on a une valeur p de 0.165772, qui est supérieure à 0.05. Par conséquent, on ne peut pas rejeter l'hypothèse nulle selon laquelle il n'y a pas d'effet significatif de "Q7" sur "Q8_encoded".
Cela signifie qu'il n'y a pas suffisamment de preuves pour conclure à une différence significative entre les groupes définis par les différentes catégories de "Q7" en termes de "Q8_encoded".

```python
from statsmodels.formula.api import ols
import statsmodels.api as sm


model = ols('Q8_encoded ~ C(Q3)', data=data_anova).fit()
anova_table = sm.stats.anova_lm(model, typ=1)
anova_table
```

Cela signifie qu'il effet significatif entre les groupes définis par les différentes catégories de "Votre formation" en termes de "Vous assistez aux cours…".
