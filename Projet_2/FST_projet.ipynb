```python
from fanalysis.mca import MCA

# instanciation
acm = MCA(row_labels=data1.index, var_labels=data1.columns)

# execution
acm.fit(data1.values)

# valuers propres
acm.eig_
```

### Informations sur les modalités

```python
# coordonnées sur le premier plan factoriel
coord_col = infos_mod[['col_coord_dim1','col_coord_dim2']]
print(coord_col)
```

```python
# nombre de lignes
n = data.shape[0]
# nombre de colonnes
p = data.shape[1]
# Contributions
col_mass = acm.c_
print(col_mass)
# normaliser les contributions (les masses) des colonnes
col_mass=col_mass/(n*p)
print(col_mass)
```

```python
# les facteurs sont centrés, vérifions avec moyenne pondérée des modalités
# sur les 2 premiers facteurs
coord_col.apply(axis=0, func=lambda x: np.sum(col_mass[0]*x))
```

### Etude de la relation entre certains traits de personnalité: ACM avec le package prince

#### Lancement de la méthode mca avec prince 

```python
#Création de l'objet mca
mca = prince.MCA()
#Lancement de l'ACM sur le jeu de données
mca=mca.fit(data2)
mca
```

#### Visualisation 

```python
# Création de la figure

fig, ax = plt.subplots(figsize=(30, 30))

#Tracage des axes
ax = plot.stylize_axis(ax)

#Affichage des profils lignes sur deux dimensions
#alpha pour la transparence 
row_coords = mca.row_coordinates(data2)
ax.scatter(
    row_coords.iloc[:, 0],
    row_coords.iloc[:, 1],
    s=10,
    label=None,
    color=plot.GRAY['dark'],
    alpha=0.6
)

#Profils colonnes
col_coords = mca.column_coordinates(data2)
x = col_coords[0]
y = col_coords[1]

#Extraction des préfixes à l'aide map et de lambda
prefixes = col_coords.index.str.split('_').map(lambda x: x[0])

#Affichage des profils colonnes avec leurs modalités
for prefix in prefixes.unique():
    mask = prefixes == prefix
    
    ax.scatter(x[mask], y[mask], s=10, label=prefix)

    for i, label in enumerate(col_coords[mask].index):
        ax.annotate(label, (x[mask][i], y[mask][i]))

#Ajout de légende
ax.legend()
            
            

ax.set_title('Row and column principal coordinates')
ei = [eig / mca.total_inertia_ for eig in mca.eigenvalues_]
ax.set_xlabel('Component {} ({:.2f}% inertia)'.format(0, 100 * ei[0]))
ax.set_ylabel('Component {} ({:.2f}% inertia)'.format(1, 100 * ei[1]))
```
